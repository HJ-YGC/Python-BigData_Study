```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Pretendard')

data = {'이름':['강감찬','홍길동','이순신','최영'],
        '점수':[50,60,80,65]}

df = pd.DataFrame(data)

# 판다스의 plot을 이용한 차트 작성
# 데이터 프레임. plot()
df.plot(kind='bar', x='이름', rot = 0, ylim=[0,100],color='y')
for i in list(df.index):
    rate = (df.loc[i]['점수']/df['점수'].sum())*100
    plt.text(x=i-0.1, y=df.loc[i]['점수']+1, s=str(df.loc[i]['점수'])+'점')
    plt.text(x=i-0.15, y=df.loc[i]['점수']/2, s=f'{rate:.1f}')

meanval = df['점수'].mean()
plt.axhline(y=meanval, color='r', linestyle='dashed')
plt.text(x=0-0.2, y=meanval+1, s=f'평균: {meanval:.1f}')

plt.xlabel('학생이름')
plt.ylabel('점수')
plt.title('학생별 시험 점수')

plt.savefig('graph01.png')
print('파일이 저장 되었습니다.')
plt.show()
```

```python
# subplot 작성

data = {'year' : [2010,2011,2012,2013,2014],
        'Sales' : [500, 800, 1200, 900, 1500],
        'Expenses': [300,400,600,500,800]
}

df2= pd.DataFrame(data)

# 서브플롯 생성
fig, axes = plt.subplots(2,2, figsize=(10,9))

# 선그래프
axes[0,0].plot(df2['year'], df2['Sales'])
axes[0,0].set_title('Sales')

# 막대그래프
axes[0,1].bar(df2['year'],df2['Expenses'])
axes[0,1].set_title('Expenses')

# 산점도 그래프
axes[1,0].scatter(df2['year'],df2['Expenses'])
axes[1,0].set_title('Sale vs Expenses')

# 히스토그램 그래프
axes[1,1].hist(df2['Sales'], bins=2)
axes[1,1].set_title('Sales Distribution')

# 그래프 간 간격 조정
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'year' : [2010,2011,2012,2013,2014],
        'Sales' : [500, 800, 1200, 900, 1500],
        'Expenses': [300,400,600,500,800]
}

df2 = pd.DataFrame(data)

# 서브 플롯 생성
fig, axes = plt.subplots(2,2, figsize=(10,9))

# 선그래프
sns.lineplot(data=df2, x='year', y='Sales', ax=axes[0,0])
axes[0,0].set_title('Sales', fontsize=15)
axes[0,0].set_xlabel('year', fontsize=9)

# 막대 그래프
sns.barplot(data=df2, x='year', y='Expenses', ax=axes[0,1], palette='Set1')

# 산점도 그래프
sns.scatterplot(data=df2, x='year', y='Expenses', ax=axes[1,0], palette='Set1')

# 히스토그램 그래프
sns.histplot(data=df2, x='Sales', ax=axes[1,1])


plt.show()
```
